### 201800839 김수연

```python
from konlpy.tag import Kkma
from collections import Counter

infile = open("소나기.txt", encoding = 'utf-8')
data = infile.read()

kkma = Kkma()
words = kkma.morphs(data)

vocab = Counter(words)

data_list = []

for n, c in vocab.most_common():
    dics = {'tag': n, 'count': c}
    if c >= 5:
        data_list.append(dics)
        
for data in data_list:
    print(data)
```

```python
from collections import Counter

infile = open("소나기.txt", encoding = 'utf-8')
data = infile.read()

kkma = Kkma()
words = kkma.morphs(data)
vocab = Counter(words)

stop_list = [',', '(', ')', '"', '"', "'", "'", '.', '!', '?',
             '은', '는', '이', '가', '를', '을', '께서', '에서', '이다', '의', '와', '과', '로', '으로', '부터', '까지', '야', '조차', '랑', '이랑', '었']

data_list = []
for n, c in vocab.most_common():
    dics = {'tag': n, 'count': c}
    if n not in stop_list:
         if c >= 5:
            data_list.append(dics)
            
for data in data_list:
    print(data)
```
